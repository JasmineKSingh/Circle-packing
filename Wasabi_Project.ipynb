```python
# _id, urlDiscogs, urlITunes, urlSpotify, urlDeezer, genres

new_df = pd.DataFrame()
new_df["ID"] = data._id
new_df["has_Discogs"] = data.urlDiscogs
new_df["has_ITunes"] = data.urlITunes
new_df["has_Spotify"] = data.urlSpotify
new_df["has_Deezer"] = data.urlDeezer
new_df["has_MusicBrainz"] = data.urlMusicBrainz
new_df["has_Allmusic"] = data.urlAllmusic
new_df["has_Wikia"] = data.urlWikia
new_df["Genres"] = data.genres
new_df["Type"] = data.type
```

```python
new_df.head()
```

```python
# Missing Values
new_df.isnull().sum()
```

```python
new_df.Genres.value_counts()
```

```python
new_df.Genres = [ast.literal_eval(gen) for gen in new_df.Genres]
```

```python
new_df = new_df.explode('Genres')
```

```python
new_df.dtypes
```

```python
# Missing Values
new_df.isnull().sum()
```

```python
new_df["ID"] = [ID.replace("ObjectId(","").replace(")","") for ID in new_df["ID"]]
```

```python
for col in new_df:
    new_df[col] = np.where(new_df[col].isnull(), "Unknown", new_df[col])
```

```python
new_df.head()
```

```python
new_df["Discogs"] = np.where(new_df["has_Discogs"]!="Unknown",1,0)
new_df["ITunes"] = np.where(new_df["has_ITunes"]!="Unknown",1,0)
new_df["Spotify"] = np.where(new_df["has_Spotify"]!="Unknown",1,0)
new_df["Deezer"] = np.where(new_df["has_Deezer"]!="Unknown",1,0)
new_df["MusicBrainz"] = np.where(new_df["has_MusicBrainz"]!="Unknown",1,0)
new_df["Allmusic"] = np.where(new_df["has_Allmusic"]!="Unknown",1,0)
new_df["Wikia"] = np.where(new_df["has_Wikia"]!="Unknown",1,0)
```

```python
del new_df["has_Discogs"]
del new_df["has_ITunes"]
del new_df["has_Spotify"]
del new_df["has_Deezer"]
del new_df["has_MusicBrainz"]
del new_df["has_Allmusic"]
del new_df["has_Wikia"]
```

```python
genres_list = set(new_df.Genres)
url_type_list = list(count_df.to_dict().keys())
type_list = set(new_df.Type)
```

```python
type_list
```

```python
keys = count_df.to_dict().keys()
genre_data = {} # data  - empty object
genre_data["name"] = "artist_data" # data  - name
genre_children = []
for genre in genres_list:
    if genre != "Unknown":
        genre_obj = {} # genre  - empty object
        genre_obj["name"] = genre # genre  - name
        type_children = []
        for type_name in type_list:
            type_obj = {} # type  - empty object
            type_obj["name"] = type_name # type  - name
            url_type_children = []
            for url_type in url_type_list:
                if url_type != "Unknown":
                    url_type_obj = {} #url_type  - empty object
                    url_type_obj["name"] = url_type #url_type  - name
                    key = (genre, type_name)
                    if key in count_df.to_dict()[url_type].keys():
                        url_type_obj["size"] = count_df.to_dict()[url_type][key] #url_type  - size
                        url_type_children.append(url_type_obj)
            type_obj["children"] = url_type_children # type  - children
            type_children.append(type_obj)
        genre_obj["children"] = type_children # genre  - children
        genre_children.append(genre_obj)
genre_data["children"] = genre_children # data  - children
```
